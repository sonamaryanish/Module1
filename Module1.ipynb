```python
'''
As a culminating project, you'll be working with a dataset from ABC company,
consisting of 458 rows and 9 columns. The company requires a comprehensive
report detailing information about their employees across various teams. Your
tasks include preprocessing the dataset, analyzing the data, and presenting your
findings graphically. Here's a breakdown of what you need to do:

'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
'''
Preprocessing:
‚óè Correct the data in the "height" column by replacing it with random
numbers between 150 and 180. Ensure data consistency and integrity
before proceeding with analysis. (1 mark)
'''

df = pd.read_excel("ABC Company_original.xlsx")
print(df.head())

random_heights = np.random.randint(150, 180, size=len(df))
df["Height"] = random_heights
df.to_excel("ABC Company_updated.xlsx", index=False)
df = pd.read_excel("ABC Company_updated.xlsx")
print(df.head())
```

```python
'''
Determine the distribution of employees across each team and calculate the percentage split 
relative to the total number of employees.
create appropriate visualizations to present your findings effectively
'''
team_counts = df["Team"].value_counts()
team_percentages = team_counts / len(df) * 100
print(team_percentages)

plt.pie(team_percentages, labels=team_percentages.index.str[:8])
plt.title("Distribution of Employees Across Teams")
plt.show()
```

```python
'''
Segregate employees based on their positions within the company
create appropriate visualizations to present your findings effectively
Data Story: Provide insights gained from the analysis, highlighting key trends, patterns, and correlations within the dataset
'''

position_count = df["Position"].value_counts()
print("\nEmployees by Position:")
print(position_count)
position_count.plot(kind="bar",title="Position Distribution",x="Position",y="Count",legend=True)
plt.show()
```

```python
'''
 Identify the predominant age group among employees. create appropriate visualizations to present 
 your findings effectively Data Story: Provide insights gained from the analysis, highlighting key trends
 , patterns, and correlations within the dataset 
 '''

age_counts = df["Age"].value_counts().sort_index()
print("\nEmployees by Age:")
print(age_counts)
age_counts.plot(kind="bar",title="Age Distribution",x="Age",y="Count",legend=True)
plt.show()
```

```python
'''
Discover which team and position have the highest salary expenditure
create appropriate visualizations to present your findings effectively 
Data Story: Provide insights gained from the analysis, highlighting key trends, patterns, and correlations within the dataset
'''
team_salary = df.groupby("Team")["Salary"].sum()
position_salary = df.groupby("Position")["Salary"].sum()
print("\nTotal Salary Expenditure by Team:")
print(team_salary)
print("\nTotal Salary Expenditure by Position:")
print(position_salary)
team_salary.plot(kind="bar",title="Total Salary Expenditure by Team",x="Team",y="Salary",legend=True)
plt.show()
position_salary.plot(kind="bar",title="Total Salary Expenditure by Position",x="Position",y="Salary",legend=True)
plt.show()
```

```python
'''
Investigate if there's any correlation between age and salary, and represent it visually.
create appropriate visualizations to present your findings effectively 
Data Story: Provide insights gained from the analysis, highlighting key trends, patterns, and correlations within the dataset

'''
corr = df["Age"].corr(df["Salary"])
print("\nCorrelation between Age and Salary:", corr)

plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs. Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

```
